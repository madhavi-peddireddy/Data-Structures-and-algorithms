```python
#Infinite Recursion(no Base condition)
def fun1():
    print(1)
    fun1();
fun1();
```

```python
count=1
def fun2():
    global count
    if count==4:
        return count
    print(count)
    count+=1
    fun2();
fun2();
```

```python
#Print name 5 times using recursion
count=0
def fun3():
    global count
    if count==5:
        return 
    print("madhavi")
    count+=1
    fun3();
fun3();
```

```python
#Print Linearly from 1 to N

count=1
def fun4():
    global count
    if count==11:
        return
    print(count)
    count+=1
    fun4();
fun4();
```

```python
#Print from N to 1

count=10
def fun5():
    global count
    if count==0:
        return
    print(count)
    count-=1
    fun5();
fun5()
```

```python
#Print name n times using Recursion(Parameters)
def fun6(i,n):
    if i>n:
        return 
    print("madhavi")
    fun6(i+1,n)
    
    
n=int(input("enter the number:"))
fun6(1,n);
```

```python
def fun7(i,N):
    if i>N:
        return 
    print(i)
    fun7(i+1,N);
    
N=int(input("enter the number"))
fun7(1,N);
```

```python
def fun8(N):
    if N==0:
        return 
    print(N)
    fun8(N-1);
N=int(input("enter the number"))
fun8(N);
```
